```python
%matplotlib inline

import pandas as pd
import tensorflow as tf
import numpy as np
import tflearn as tflearn
from clean_data import clean_data
```

```python
df = pd.read_csv('../train.csv')
labels = pd.get_dummies(df.Survived)
df = df.drop('Survived', 1)
df.head()
```

```python
test = pd.read_csv('../test.csv')
df = clean_data(df)
test = clean_data(test)
df
```

```python
labels = np.array(labels, dtype=np.float32)
```

```python
data = df
net = tflearn.input_data(shape=[None, 9]) 
net = tflearn.fully_connected(net, 7)
net = tflearn.fully_connected(net, 7)
net = tflearn.fully_connected(net, 7)
net = tflearn.fully_connected(net, 2, activation='softmax') 
net = tflearn.regression(net)
model = tflearn.DNN(net)
model.fit(data, labels, n_epoch=20, batch_size=16, show_metric=True)
```

```python
pred = model.predict(test)
pred = pd.DataFrame(pred)
pred = pred.drop(0, 1)
```

```python
end = pd.read_csv('../test.csv')
end['Survival_chance'] = pred
state = []
for i in end['Survival_chance']:
    if i > 0.5:
        state.append('Alive')
    else:
        state.append('Dead')
state
end['State'] = state
```
